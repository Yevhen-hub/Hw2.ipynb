```python
import pandas as pd

df = pd.read_csv("2017_jun_final - 2017_jun_final.csv", encoding="utf-8")
df.head()
df.shape
df.dtypes
df.isnull().sum() / len(df)

df_cols = df.copy()
lang_col = "Мова програмування"
cols_with_na = df_cols.columns[df_cols.isnull().any()]
cols_to_drop = [c for c in cols_with_na if c != lang_col]
df_cols = df_cols.drop(columns=cols_to_drop)
df_cols.isnull().sum() / len(df_cols)

df_no_na = df.dropna()
df_no_na.shape

python_data = df_no_na[df_no_na["Мова програмування"] == "Python"]
python_data.shape

grouped = python_data.groupby("Посада")
salary_agg = grouped.agg({"Зарплата на місяць": ["min", "max"]})
salary_agg.columns = ["min_salary", "max_salary"]
salary_agg = salary_agg.reset_index()

def fill_avg_salary(row):
    return (row["min_salary"] + row["max_salary"]) / 2

salary_agg["avg"] = salary_agg.apply(fill_avg_salary, axis=1)
salary_agg["avg"].describe()

salary_agg.to_csv("python_salary_by_position.csv", index=False, encoding="utf-8")
```
