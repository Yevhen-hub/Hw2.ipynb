```python
import pandas as pd
import numpy as np
import glob

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 300)
pd.set_option("display.max_colwidth", None)

path = glob.glob("/content/**/*.csv", recursive=True)[0]
df = pd.read_csv(path)

salary_col = [c for c in df.columns if "зарплата" in c.lower()][0]
lang_col = [c for c in df.columns if "мова" in c.lower()][0]
position_col = [c for c in df.columns if "посад" in c.lower()][0]

cols_to_drop = [c for c in df.columns if c != lang_col and df[c].isnull().any()]
df_no_na_cols = df.drop(columns=cols_to_drop)

df_clean = df.dropna()
df_clean[salary_col] = pd.to_numeric(df_clean[salary_col], errors="coerce")
df_clean = df_clean.dropna(subset=[salary_col])

python_data = df_clean[df_clean[lang_col] == "Python"]

grouped = python_data.groupby(position_col)[salary_col].agg(["min", "max"])

def fill_avg_salary(row):
    return (row["min"] + row["max"]) / 2

grouped["avg"] = grouped.apply(fill_avg_salary, axis=1)

grouped.to_csv("python_salary_by_position.csv", encoding="utf-8")
```
